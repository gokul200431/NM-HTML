```python
class ColorPaletteGenerator:
    def __init__(self):
        self.palettes = {
            "modern": {
                "primary": "#4f46e5",
                "secondary": "#6366f1",
                "background": "#0f172a",
                "text": "#f8fafc"
            },
            "ocean": {
                "primary": "#0ea5e9",
                "secondary": "#0284c7",
                "background": "#082f49",
                "text": "#e0f2fe"
            },
            "forest": {
                "primary": "#16a34a",
                "secondary": "#15803d",
                "background": "#052e16",
                "text": "#dcfce7"
            }
        }

    def get_palette(self, style):
        return self.palettes.get(style, self.palettes["modern"])
```

```python
class HTMLPageBuilder:
    def __init__(self):
        self.sections = []

    def detect_sections(self, description):
        keywords = ["hero", "features", "about", "portfolio", "contact", "testimonials"]
        for word in keywords:
            if word in description.lower():
                self.sections.append(word)
        return self.sections

    def generate_html(self, title, description, palette):
        self.detect_sections(description)

        html = f"""
        <!DOCTYPE html>
        <html>
        <head>
            <title>{title}</title>
            <style>
                body {{
                    background-color: {palette['background']};
                    color: {palette['text']};
                    font-family: Arial;
                }}
                h1 {{
                    color: {palette['primary']};
                }}
            </style>
        </head>
        <body>
            <h1>{title}</h1>
        """

        for section in self.sections:
            html += f"<section><h2>{section.title()}</h2><p>Generated {section} section</p></section>"

        html += "</body></html>"

        return html
```

```python
import gradio as gr

palette_generator = ColorPaletteGenerator()
page_builder = HTMLPageBuilder()

def generate_page(title, description, style):
    palette = palette_generator.get_palette(style)
    html = page_builder.generate_html(title, description, palette)
    return html

interface = gr.Interface(
    fn=generate_page,
    inputs=[
        gr.Textbox(label="Page Title"),
        gr.Textbox(label="Description"),
        gr.Dropdown(["modern", "ocean", "forest"], label="Color Style")
    ],
    outputs=gr.HTML(label="Generated Page"),
    title="HTML Quick Styler"
)

interface.launch()
```
