```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_id = "ibm-granite/granite-3.3-2b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto"
)

print("Model loaded successfully")
```

```python
def generate_text(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=100,
        temperature=0.7,
        do_sample=True
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)

print(generate_text("Write a simple HTML hero section for a portfolio website."))
```

```python
class HTMLPageBuilder:
    def __init__(self):
        pass

    def detect_sections(self, description):
        description = description.lower()
        sections = []

        keywords = {
            "hero": ["hero", "welcome", "banner"],
            "features": ["feature", "features", "services"],
            "about": ["about", "bio", "company overview"],
            "portfolio": ["portfolio", "projects", "gallery"],
            "testimonials": ["testimonial", "reviews"],
            "pricing": ["pricing", "plans"],
            "contact": ["contact", "form"],
            "footer": ["footer"]
        }

        for section, words in keywords.items():
            if any(word in description for word in words):
                sections.append(section)

        return sections

    def build_section(self, section):
        templates = {
            "hero": """
<section class="hero">
    <h1>Welcome to My Website</h1>
    <p>Your success starts here.</p>
    <button>Get Started</button>
</section>
""",
            "features": """
<section class="features">
    <h2>Features</h2>
    <div class="feature-grid">
        <div class="card">Feature One</div>
        <div class="card">Feature Two</div>
        <div class="card">Feature Three</div>
    </div>
</section>
""",
            "about": """
<section class="about">
    <h2>About Us</h2>
    <p>This is a professional description about the company or individual.</p>
</section>
""",
            "portfolio": """
<section class="portfolio">
    <h2>Our Work</h2>
    <div class="gallery">
        <div class="card">Project 1</div>
        <div class="card">Project 2</div>
        <div class="card">Project 3</div>
    </div>
</section>
""",
            "testimonials": """
<section class="testimonials">
    <h2>Testimonials</h2>
    <div class="card">"Great service!"</div>
</section>
""",
            "pricing": """
<section class="pricing">
    <h2>Pricing Plans</h2>
    <div class="card">Basic Plan</div>
    <div class="card">Pro Plan</div>
</section>
""",
            "contact": """
<section class="contact">
    <h2>Contact Us</h2>
    <form>
        <input type="text" placeholder="Your Name" required>
        <input type="email" placeholder="Your Email" required>
        <textarea placeholder="Your Message"></textarea>
        <button type="submit">Send Message</button>
    </form>
</section>
""",
            "footer": """
<footer class="footer">
    <p>© 2026 Your Company. All rights reserved.</p>
</footer>
"""
        }

        return templates.get(section, "")

    def build_page(self, title, description):
        sections = self.detect_sections(description)

        html_sections = ""
        for section in sections:
            html_sections += self.build_section(section)

        full_html = f"""
<!DOCTYPE html>
<html>
<head>
    <title>{title}</title>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
</head>
<body>
{html_sections}
</body>
</html>
"""
        return full_html
```

```python
builder = HTMLPageBuilder()

html_output = builder.build_page(
    "Modern Portfolio Website",
    "Hero section with welcome message, features list, about section, portfolio gallery, testimonials, contact form and footer"
)

print(html_output)
```

```python
class CSSGenerator:
    def generate_css(self, palette):
        return f"""
:root {{
    --primary: {palette['primary']};
    --secondary: {palette['secondary']};
    --accent: {palette['accent']};
    --background: {palette['background']};
    --surface: {palette['surface']};
    --text-primary: {palette['text_primary']};
    --text-secondary: {palette['text_secondary']};
    --border: {palette['border']};
}}

body {{
    margin: 0;
    font-family: Arial, sans-serif;
    background-color: var(--background);
    color: var(--text-primary);
}}

section {{
    padding: 60px 20px;
    text-align: center;
}}

h1, h2 {{
    color: var(--primary);
}}

button {{
    padding: 10px 20px;
    border: none;
    background-color: var(--primary);
    color: white;
    cursor: pointer;
    transition: 0.3s ease;
}}

button:hover {{
    background-color: var(--secondary);
}}

.card {{
    background-color: var(--surface);
    padding: 20px;
    margin: 10px;
    border-radius: 8px;
    border: 1px solid var(--border);
    display: inline-block;
}}

.feature-grid, .gallery {{
    display: flex;
    justify-content: center;
    flex-wrap: wrap;
}}

input, textarea {{
    padding: 10px;
    margin: 10px;
    width: 250px;
}}

footer {{
    padding: 20px;
    background-color: var(--surface);
}}
"""
```

```python
palette_gen = ColorPaletteGenerator()
css_gen = CSSGenerator()

palette = palette_gen.get_palette("modern")
css_output = css_gen.generate_css(palette)

print(css_output)
```

```python
class FinalPageBuilder:
    def __init__(self):
        self.html_builder = HTMLPageBuilder()
        self.palette_generator = ColorPaletteGenerator()
        self.css_generator = CSSGenerator()

    def build_complete_page(self, title, description, style):
        palette = self.palette_generator.get_palette(style)
        css = self.css_generator.generate_css(palette)
        raw_html = self.html_builder.build_page(title, description)

        # Inject CSS inside <head>
        final_html = raw_html.replace(
            "</head>",
            f"<style>{css}</style></head>"
        )

        return final_html
```

```python
final_builder = FinalPageBuilder()

complete_html = final_builder.build_complete_page(
    "Modern Portfolio Website",
    "Hero section with welcome message, features list, about section, portfolio gallery, testimonials, contact form and footer",
    "modern"
)

print(complete_html)
```

```python
import gradio as gr

final_builder = FinalPageBuilder()

def generate_page(title, description, style):
    return final_builder.build_complete_page(title, description, style)

with gr.Blocks() as app:
    gr.Markdown("# HTML Quick Styler")

    with gr.Row():
        title_input = gr.Textbox(label="Page Title")
        style_dropdown = gr.Dropdown(
            choices=["modern", "vibrant", "ocean", "forest", "sunset", "luxury", "creative", "minimal"],
            value="modern",
            label="Color Style"
        )

    description_input = gr.Textbox(
        label="Page Description",
        lines=4,
        placeholder="Describe sections like hero, features, about, portfolio, contact..."
    )

    generate_button = gr.Button("Generate HTML")

    output_html = gr.Code(label="Generated HTML", language="html")

    generate_button.click(
        fn=generate_page,
        inputs=[title_input, description_input, style_dropdown],
        outputs=output_html
    )

app.launch(share=True)
```

```python
import gradio as gr

final_builder = FinalPageBuilder()
palette_gen = ColorPaletteGenerator()

def generate_page(title, description, style):
    return final_builder.build_complete_page(title, description, style)

def show_palette(style):
    return palette_gen.get_palette(style)

with gr.Blocks() as app:
    gr.Markdown("# HTML Quick Styler")

    with gr.Tabs():

        # TAB 1 — Generate Page
        with gr.Tab("Generate Page"):
            title_input = gr.Textbox(label="Page Title")
            style_dropdown = gr.Dropdown(
                choices=["modern", "vibrant", "ocean", "forest", "sunset", "luxury", "creative", "minimal"],
                value="modern",
                label="Color Style"
            )
            description_input = gr.Textbox(
                label="Page Description",
                lines=4,
                placeholder="Describe sections like hero, features, about, portfolio, contact..."
            )
            generate_button = gr.Button("Generate HTML")
            output_html = gr.Code(label="Generated HTML", language="html")

            generate_button.click(
                fn=generate_page,
                inputs=[title_input, description_input, style_dropdown],
                outputs=output_html
            )

        # TAB 2 — Live Preview
        with gr.Tab("Live Preview"):
            preview_html = gr.HTML()
            preview_button = gr.Button("Generate Preview")

            preview_button.click(
                fn=generate_page,
                inputs=[title_input, description_input, style_dropdown],
                outputs=preview_html
            )

        # TAB 3 — Color Palettes
        with gr.Tab("Color Palettes"):
            palette_style = gr.Dropdown(
                choices=["modern", "vibrant", "ocean", "forest", "sunset", "luxury", "creative", "minimal"],
                value="modern",
                label="Select Palette"
            )
            palette_output = gr.JSON(label="Palette Colors")

            palette_style.change(
                fn=show_palette,
                inputs=palette_style,
                outputs=palette_output
            )

        # TAB 4 — Templates
        with gr.Tab("Templates"):
            gr.Markdown("## Quick Templates")

            def portfolio_template():
                return "Hero section with welcome message, features section, about section, portfolio gallery, testimonials, contact form and footer"

            def saas_template():
                return "Hero banner with product message, features section with icons, pricing plans table, testimonials and call to action"

            def company_template():
                return "Sticky navigation header, company overview, services grid, team members section, contact form and footer"

            portfolio_btn = gr.Button("Portfolio Website")
            saas_btn = gr.Button("SaaS Landing Page")
            company_btn = gr.Button("Company Website")

            portfolio_btn.click(
                fn=portfolio_template,
                outputs=description_input
            )

            saas_btn.click(
                fn=saas_template,
                outputs=description_input
            )

            company_btn.click(
                fn=company_template,
                outputs=description_input
            )

            gr.Markdown("""
### Available Sections:
- Hero
- Features
- About
- Portfolio
- Testimonials
- Pricing
- Contact
- Footer

Include these keywords in description to auto-generate sections.
""")

        # TAB 5 — Guide
        with gr.Tab("Guide"):
            gr.Markdown("""
### How to Use:
1. Enter a page title.
2. Describe the sections you want.
3. Choose a color style.
4. Click Generate.
5. Use Live Preview to see output.

Example description:
Hero section, features list, about section, portfolio gallery, testimonials, contact form, footer
""")

app.launch(share=True)
```
