```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

model_name = "ibm-granite/granite-3.3-2b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto"
)

generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)
```

```python
class ColorPaletteGenerator:

    def __init__(self):
        self.palettes = {
            "modern": {
                "primary": "#6366f1",
                "secondary": "#8b5cf6",
                "accent": "#22c55e",
                "background": "#0f172a",
                "surface": "#1e293b",
                "text_primary": "#ffffff"
            }
        }

    def get_palette(self, style):
        return self.palettes.get(style, self.palettes["modern"])
```

```python
import gradio as gr

def generate_page(title, description):

    prompt = f"""
Create a modern HTML webpage.

Title: {title}

Description: {description}

Return HTML code.
"""

    result = generator(
        prompt,
        max_new_tokens=400
    )[0]["generated_text"]

    return result


with gr.Blocks() as app:

    gr.Markdown("# HTML Quick-Styler")

    title = gr.Textbox(label="Page Title")

    description = gr.Textbox(label="Page Description")

    btn = gr.Button("Generate")

    output = gr.Code()

    btn.click(
        generate_page,
        inputs=[title, description],
        outputs=output
    )

app.launch(share=True)
```
